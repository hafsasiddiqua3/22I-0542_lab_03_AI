```python
menu = {"lolipop": "20", "watermelon": "30", "popsicle": "25"}

print(" ~~~~~~~~~~~Welcome ~~~~~~~~~~~")
print("            Menu :)")
print("Enter the item number you want to order (or 'q' to quit)")

num = 0
items=[]
prices=[]
for key, value in menu.items():
    num += 1
    print(num, key, value)

while True:
    choice_input = input("Enter your choice: ")
    if choice_input == "q":
        break
    choice = int(choice_input)
    if choice > len(menu) or choice <= 0:
        print("Invalid choice, try again.")
    else:
        item = list(menu.keys())[choice - 1]
        price = menu[item]
        print(f"You ordered {item} which costs {price} dollars.")

        items.append(item)
        prices.append(price)
for i in range(len(items)):
  print(items[i]," " ,prices[i])

print("~~~~~~~~~~~GOODBYE :))))2  ~~~~~~~~~~~")

```

```python
important=[]
promotions=[]
spam=[]
i=0
emails=['this is an important email','this email contain promotions','this is a spam email','this is a spam email']
def classify_email(x):
 for email in x:
  if 'important' in email.lower():
    important.append(email)
  if 'promotions' in email.lower():
      promotions.append(email)
  if 'spam' in email.lower():
    spam.append(email)

 return important,promotions,spam

classify_email(emails)
print(important)
print(promotions)
print(spam)
```

```python
import pandas as pd
url="https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/vgsalesGlobale.csv"
dataframe=pd.read_csv(url)
df=pd.DataFrame(dataframe)
df
```

```python
global_saless=df['Global_Sales'].sum()

print("total global sales ",global_saless)
#game with highest global sales
max_sales = df['Global_Sales'].max()

# Retrieve the 'Name' column where 'Global_Sales' is equal to the maximum value
max_sales = df.loc[df['Global_Sales'] == max_sales, 'Name']

print("name of max sales game ",max_sales)
#average sales for NA_Sales
NA_Sales_average=df['NA_Sales'].mean()
#average sales for EU_Sales
EU_Sales_average=df['EU_Sales'].mean()
#average sales for JP_Sales
JP_Sales_average=df['JP_Sales'].mean()
print('NA_Sales_average ',NA_Sales_average)
print('EU_Sales_average',EU_Sales_average)
print('JP_Sales_average',JP_Sales_average)
top_genres_global_sales = df.groupby('Genre')['Global_Sales'].sum().nlargest(5)
print(top_genres_global_sales)
total_sales_by_genre_region = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
most_popular_genre_in_NA = total_sales_by_genre_region['NA_Sales'].idxmax()
print("most_popular_genre_in_NA")
```

```python
last_10_rows = df.tail(10)
last_10_rows
```

```python
top_3_platforms_global_sales = df.groupby('Platform')['Global_Sales'].sum().nlargest(3)
top_3_platforms_global_sales
```

```python
genre_market_share = (df.groupby('Genre')['Global_Sales'].sum() / df['Global_Sales'].sum()) * 100
genre_market_share
```
